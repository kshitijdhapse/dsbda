```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read the dataset
df = pd.read_csv("Hydra-Movie-Scrape.csv")

df.head()
```

```python
#Ithe majhyakade ya assignment saathi kuthlach dataset match hot nahiye . Movie
# dataset ch aahet , pan question madhe languages vagaire ni group karaychet
# je ki yaa dataset made nahiye
#mhnaun mi question soy anusar badallet




# m. Create data subsets for different years
subsets_by_year = df.groupby('Year')

subsets_by_year.head()
```

```python
# n. Merge two subsets
subset1 = subsets_by_year.get_group(2017)
subset2 = subsets_by_year.get_group(2008)
merged_subset = pd.concat([subset1, subset2])

merged_subset.head()
```

```python



#  Sort Data using customer ratings
sorted_df = df.sort_values(by='Runtime', ascending=False)

sorted_df.head()
```

```python




# p. Transposing Data
transposed_df = df.transpose()

transposed_df.head()
```

```python



# q. Melting Data to long format
melted_df = pd.melt(df, id_vars=['IMDB ID', 'Title', 'Year'], var_name='Attribute', value_name='Value')

melted_df.head()
```

```python


# r. Casting data to wide format
casted_df = melted_df.pivot_table(index=['IMDB ID', 'Title', 'Year'], columns='Attribute', values='Value', aggfunc='first').reset_index()

casted_df.head()
```
