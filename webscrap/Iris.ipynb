```python
def remove_outlier(column):
    Q1=column.quantile(0.25)
    Q3=column.quantile(0.75)
    IQR=Q3-Q1
    threshold=1.5*IQR
    maskcol=(column<Q1-threshold) | (column>Q3+threshold)
    return column[~maskcol]
```

```python
for colu in colus:
    df_clean[colu]=remove_outlier(df[colu])
```

```python
# y=np.array(y)
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
y=le.fit_transform(df_clean['species'])
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.7,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scalertrain = StandardScaler()
```

```python

x_train_scaled=scalertrain.fit_transform(x_train)
x_test_scaled=scalertrain.transform(x_test)
x_train_scaled.shape
```

```python
y_train=np.array(y_train).reshape(-1,1)
y_test=np.array(y_test).reshape(-1,1)
y_train.shape
```

```python
np.isnan(x_train_scaled).sum()
```

```python
from sklearn.linear_model import LogisticRegression

model1=LogisticRegression();
model1.fit(x_train_scaled,y_train)
```

```python
y_pred=model1.predict(x_test_scaled)
y_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
a=accuracy_score(y_test,y_pred)
a
```

```python
from sklearn.naive_bayes import GaussianNB
model2=GaussianNB()
model2.fit(x_train_scaled,y_train)
```

```python
y_pred2=model2.predict(x_test_scaled)
y_pred2
```

```python
a=accuracy_score(y_test,y_pred2)
a
```

```python
from sklearn.neighbors import KNeighborsClassifier
model3 = KNeighborsClassifier()
```

```python
model3.fit(x_train_scaled,y_train)
```

```python
y_pred3=model3.predict(x_test_scaled)
y_pred3
```

```python
a=accuracy_score(y_test,y_pred3)
a
```

```python
from sklearn.tree import DecisionTreeClassifier
model4=DecisionTreeClassifier()
model4.fit(x_train_scaled,y_train)
```
