```python

# g. Create data subsets for different species


setosa_subset = df[df['Species'] == 'Iris-setosa']
versicolor_subset = df[df['Species'] == 'Iris-versicolor']
virginica_subset = df[df['Species'] == 'Iris-virginica']
```

```python
# h. Merge two subsets (Merge setosa_subset and versicolor_subset)
merged_subset = pd.concat([setosa_subset, versicolor_subset], axis=0)

merged_subset.head()
```

```python
# i. Sort Data Petal Length
sorted_df = df.sort_values(by='PetalLengthCm')

sorted_df
```

```python
# j. Transposing Data
transposed_df = df.transpose()

transposed_df
```

```python
# k. Melting Data to long format
melted_df = pd.melt(df, id_vars=['Id', 'Species'], var_name='Measurement', value_name='Value')

melted_df
```

```python
# l. Casting data to wide format (assuming you want to pivot based on Species)
wide_df = df.pivot_table(index='Id', columns='Species')

wide_df
```
